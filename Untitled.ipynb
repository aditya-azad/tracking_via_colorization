```python
images_paths = sorted(glob.glob("../dataset/tracking/davis/train/JPEGImages/480p/dogs-jump/*"),key=lambda x: int(os.path.basename(x).split(".")[0]))
```

```python
images_paths[0]
```

```python
num_ref = 1
alpha = 0.7



ref = cv2.imread(images_paths[0])
ho,wo = ref.shape[:2]

vid = cv2.VideoWriter(
                "output_dog.mp4",
                cv2.VideoWriter_fourcc(*"MJPG"),
                4,
                (wo,ho),
            )

detections = predictor(ref)["instances"].pred_masks.int()[2]
ref_mask = Image.fromarray(detections.cpu().numpy().copy()).resize((32,32))

for i,path in enumerate(images_paths[1:]):
    tar = cv2.imread(path)
    inputs = transform_training([ref,tar])
    sim = engine.model(inputs.to("cuda"),num_ref)
    
    ref_mask_t = ToTensor()(ref_mask)
    ref_mask_t = ref_mask_t.to("cuda").float()
    sim = sim.to("cuda")
    
    num_classes, h, w = ref_mask_t.shape
    # ref_labels = ref_mask_t.permute([0, 2, 1, 3])
    ref_mask_t = ref_mask_t.reshape(-1, num_classes, h * w * num_ref)
#     print(ref_mask_t.shape,sim.shape)
    predicted_labels = torch.matmul(ref_mask_t, sim)
    
    predicted_labels = predicted_labels.reshape(32,32).cpu().detach().numpy()
    
    predicted_labels_o = (Image.fromarray(predicted_labels)).resize((wo,ho),4)
    mask = np.stack([np.array(predicted_labels_o).astype("float")]*3,axis=2) 
    o = mask* np.array([0,255,0]) * alpha +  tar * (mask) * (1-alpha) + tar * (1-mask)
#     show(o.astype("uint8"),num=i)
    
    ref = tar
    ref_mask = predicted_labels
    
#     if i==1:
#         break
    vid.write(o[...,::-1].astype("uint8"))
```
